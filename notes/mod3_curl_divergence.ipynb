```python
import numpy as np
import matplotlib.pyplot as plt
import math

import sympy as sym

sym.init_printing()
```

```python
x, y, z = sym.symbols('x y z')
```

image.png

```python
def curl_vec(P, Q, R):
  dRy = sym.diff(R, y)
  dQz = sym.diff(Q, z)
  dRx = sym.diff(R, x)
  dPz = sym.diff(P, z)
  dQx = sym.diff(Q, x)
  dPy = sym.diff(P, y)
  curl = [(dRy - dQz), -(dRx - dPz), (dQx-dPy)]
  return curl

def div_vec(P, Q, R):
  dPx = sym.diff(P, x)
  dQy = sym.diff(Q, y)
  dRz = sym.diff(R, z)
  div = dPx + dQy + dRz
  return div
```

## Curl

$$ curl \ F(x,y,z) = \nabla \times F(x,y,z) $$  

$$ curl \ F(x,y,z) = \begin{vmatrix}
i & j & k \\
\frac{\partial}{\partial x} & \frac{\partial}{\partial y} & \frac{\partial}{\partial z} \\
f & g & h \\
\end{vmatrix} $$  

## Diveregence  
  
The divergence of 
$$ F(x, y, z) = \left \langle f(x,y,z), g(x, y, z), h(x, y, z) \right \rangle $$
  

$$ div F(x, y, z) = \nabla \cdot F $$  


$$ div F(x,y,z) = \frac{\partial f}{\partial x} + \frac{\partial g}{\partial y} + \frac{\partial h}{\partial z} $$  

For a 2D vector field
$$ div F(x,y) = \frac{\partial f}{\partial x} + \frac{\partial g}{\partial y} $$

## Plotting


```python
# Work in Progress

ax = plt.figure(figsize=(10,10)).add_subplot(projection='3d')
lim = [-2, 2]
N = 5

# Make the grid
X, Y, Z = np.meshgrid(np.linspace(lim[0], lim[1], N),
                      np.linspace(lim[0], lim[1], N),
                      np.linspace(lim[0], lim[1], N),)

# Make the direction data for the arrows
u = Y*X**6
v = X*Z**5
w = Z*Y**4

c = np.sqrt(v ** 2 + u ** 2 + w ** 2)
c = (c.ravel() - c.min()) / c.ptp()
  # Repeat for each body line and two head lines
c = np.concatenate((c, np.repeat(c, 2)))
  # Colormap
c = plt.cm.coolwarm(c)
ax.quiver(X, Y, Z, u, v, w, colors=c, length=0.32, normalize=True)

plt.show()
```
