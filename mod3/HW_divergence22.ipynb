## Divergence Theorem

$$ div \vec{F} (x, y, z) = \nabla \cdot \vec{F} $$

$$ div \vec{F}(x, y, z) = \frac{\partial P}{\partial x} + \frac{\partial Q}{\partial y} + \frac{\partial R}{\partial z} $$

image.png

```python
def divF(P, Q, R):
  dPx = sym.diff(P, x)
  dQy = sym.diff(Q, y)
  dRz = sym.diff(R, z)

  return dPx + dQy + dRz
```

### Question 3:

Calculate $ \iint_S \vec{F} \cdot d\bar{S} $ where

$ \vec{F} = \left \langle e^x\sin y, e^x \cos y, yz^2 \right \rangle $  

$ S $ is the surface of the box bounded by the planes $ x = 5, y = 4, z = 1 $, and the coordinate planes.

```python
P, Q, R = [sym.E**x*sym.sin(y), sym.E**x*sym.cos(y), y*z**2]
div_F = divF(P, Q, R)
div_F
```

```python
intLim = [(x, 0, 5), (y, 0, 4), (z, 0, 1)]
sol = sym.integrate(div_F, intLim[0], intLim[1], intLim[2])
sol
```

### Question 4:

Calculate $ \iint_S \vec{F} \cdot d \bar{S} $ where  
$ \vec{F} = \left \langle 3xy^2, xe^z, z^3 \right \rangle $  
$ S $ is the surface of the solid bounded by the cylinder $ y^2 + z^2 = 1 $ and the planes $ x = -4 $ and $ x = -1 $

```python
P, Q, R = [3*x*y**2, x*sym.E**z, z**3]
div_F = divF(P, Q, R)
div_F = div_F.factor()
div_F
```

```python
r, theta = sym.symbols('r theta')
intLim = [(x, -4, -1), (r, 0 , 1), (theta, 0, 2*sym.pi)]

sol = sym.integrate(div_F.subs(y**2 + z**2, r**2)*r, intLim[0], intLim[1], intLim[2])
sol
```

### Question 5:

Calculate $ \iint_S \vec{F} \cdot d\vec{S} $ where  

$ \vec{F} = \left \langle 4x^3z, 4y^3z, 3z^4 \right \rangle $  
  
$ S $ is the surface of the solid bounded by the hemispheres $ z = \sqrt{25 - x^2 - y^2}, z = \sqrt{4-x^2-y^2} $ and the plane $ z = 0 $  

$ dV = \rho ^2 \sin(\phi) d \rho d \phi d \theta $

```python
P, Q, R = [4*x**3*z, 4*y**3*z, 3*z**4]

rho, phi, theta = sym.symbols('rho phi theta')
dV = rho**2*sym.sin(phi)
```

```python
div_F = divF(P, Q, R)
div_F = div_F.factor()
div_F
```

```python
div_F = div_F.subs(x**2 + y**2 + z**2, rho**2)
div_F = div_F.subs(z, rho*sym.cos(phi))
div_F
```

```python
intLim = [(rho, 2, 5), (phi, 0, sym.pi/2), (theta, 0, 2*sym.pi)]
sym.integrate(div_F*dV, intLim[0], intLim[1], intLim[2])
```
