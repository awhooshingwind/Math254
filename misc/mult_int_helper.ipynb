```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
import plotly.express as px

init_printing()
```

```python
x,y = symbols('x y')
x, y 
```

```python
func = 2*x + 4*y
int_y = integrate(func,(y, x**2, 4))
int_x = integrate(int_y, (x, 1,2))
int_x
```

```python
lam_f = lambdify((x,y), func, 'numpy')
x_vals = np.linspace(1, 2, 5)
y_vals = np.linspace(1,4,5)
X,Y = np.meshgrid(x_vals, y_vals)
Z = lam_f(X,Y)
#print(plot_vals)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z)
x_domain = [1, 2]
y_domain = np.square(x_domain)
#ax.plot(x_domain, y_domain, zs=0, zdir='z', label='curve in (x, y)')
plt.show()

plt.plot(x_domain, y_domain)
plt.plot()
```

```python
import plotly.graph_objects as go

fig = go.Figure(go.Surface(
    contours = {
        "x": {"show": True, "start": 1, "end": 10, "size": 1, "color":"white"},
        "z": {"show": True, "start": 0, "end": 8, "size": 1}
    },
    x = x_vals,
    y =y_vals,
    z = Z
    ))
fig.update_layout(
        scene = {
            "xaxis": {"nticks": 10},
            "zaxis": {"nticks": 10},
            'camera_eye': {"x": 0, "y": -1, "z": 0.5},
            "aspectratio": {"x": 1, "y": 1, "z": 0.2}
        })
fig.show()
```
